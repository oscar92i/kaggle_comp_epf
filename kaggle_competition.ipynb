```python
from sklearn.preprocessing import LabelEncoder

# Fill missing values
for column in df_train.columns:
    if df_train[column].dtype == 'object':
        df_train[column].fillna(df_train[column].mode()[0], inplace=True)
    elif df_train[column].dtype == 'bool':
        df_train[column].fillna(df_train[column].mode()[0], inplace=True)
    else:
        df_train[column].fillna(df_train[column].median(), inplace=True)

# Encode categorical variables
label_encoders = {}
for column in df_train.select_dtypes(include=['object']).columns:
    if column != 'auctionId':
        le = LabelEncoder()
        df_train[column] = le.fit_transform(df_train[column])
        label_encoders[column] = le
for column in df_train.select_dtypes(include=['bool']).columns:
    df_train[column] = df_train[column].astype(int)

# Display the cleaned data
df_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Fill missing values
for column in df_test.columns:
    if df_test[column].dtype == 'object':
        df_test[column].fillna(df_test[column].mode()[0], inplace=True)
    elif df_test[column].dtype == 'bool':
        df_test[column].fillna(df_test[column].mode()[0], inplace=True)
    else:
        df_test[column].fillna(df_test[column].median(), inplace=True)

# Encode categorical variables
label_encoders = {}
for column in df_test.select_dtypes(include=['object']).columns:
    if column != 'auctionId':
        le = LabelEncoder()
        df_test[column] = le.fit_transform(df_test[column])
        label_encoders[column] = le
for column in df_test.select_dtypes(include=['bool']).columns:
    df_test[column] = df_test[column].astype(int)

# Display the cleaned data
df_test.head()
```

```python
df_train['timeStamp'] = pd.to_datetime(df_train['timeStamp'], unit='s')
df_train['hour'] = df_train['timeStamp'].dt.hour
df_train['day_of_week'] = df_train['timeStamp'].dt.dayofweek
df_train['day_of_month'] = df_train['timeStamp'].dt.day
df_train['month'] = df_train['timeStamp'].dt.month
df_train['year'] = df_train['timeStamp'].dt.year
df_train['auctionId'] = df_train['auctionId'].astype(str)
# Drop the original timeStamp column
df_train.drop('timeStamp', axis=1, inplace=True)

# Display the engineered features
df_train.head()
```

```python
df_test['timeStamp'] = pd.to_datetime(df_test['timeStamp'], unit='s')
df_test['hour'] = df_test['timeStamp'].dt.hour
df_test['day_of_week'] = df_test['timeStamp'].dt.dayofweek
df_test['day_of_month'] = df_test['timeStamp'].dt.day
df_test['month'] = df_test['timeStamp'].dt.month
df_test['year'] = df_test['timeStamp'].dt.year
df_test['auctionId'] = df_test['auctionId'].astype(str)
# Drop the original timeStamp column
df_test.drop('timeStamp', axis=1, inplace=True)

# Display the engineered features
df_test.head()
```
